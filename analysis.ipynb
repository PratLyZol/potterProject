```python
import pandas as pd     
cancer = pd.read_csv("Cancer.csv")
```

```python
cancer.drop("id", inplace=True, axis=1)
cancer
```

```python
cancer.drop("Unnamed: 32", inplace=True, axis=1)
```

```python
cancer
```

```python
from sklearn.model_selection import train_test_split
X = cancer.drop("diagnosis", axis=1)
y = cancer["diagnosis"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=69)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
max = 0.0
maxK = 0
# Plotting Accuracy
accuracies = []
for i in range(30):
    knn = KNeighborsClassifier(n_neighbors=(i+1), p = 5)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = np.mean(y_pred == list(y_test))
    if accuracy > max:
        max = accuracy
        maxK = i+1
    accuracies.append(accuracy)
plt.plot(range(1, 31), accuracies)
plt.xlabel("n_neighbors")
plt.ylabel("accuracy")
plt.title("Minkowski Distance Accuracy")
```

```python
from mlxtend.plotting import plot_decision_regions
from sklearn.decomposition import PCA 
from sklearn.metrics import accuracy_score
from sklearn.inspection import DecisionBoundaryDisplay
# Using a PCA to visualize how algorithm makes decisions
pca = PCA(n_components=2)
X = pca.fit_transform(X_train)
y = y_train.replace('M', 2)
y = y.replace('B', 1)
y = y.to_numpy()
KNN = KNeighborsClassifier(n_neighbors=6, p = 2)
KNN = KNN.fit(X, y)
x_test = pca.fit_transform(X_test)
y_true2 = y_test.replace('M', 2)
y_true2 = y_true2.replace('B', 1)
predicted = KNN.predict(x_test)
print(accuracy_score(y_true2, predicted))
disp = DecisionBoundaryDisplay.from_estimator(
    KNN, X, response_method="predict", alpha=0.5
)
disp.ax_.scatter(X[:, 0], X[:, 1], c = y, edgecolors="k")
disp.ax_.legend()
plt.title("Decision Boundary for KNN")
plt.show()
# abcd
```

```python
important_features = []
explained_variance_ratio = pca.explained_variance_ratio_
for i in range(len(explained_variance_ratio)):
    if explained_variance_ratio[i] > 0.1:
        important_features.append(X[i])
print(important_features)
```

```python
from sklearn import metrics
# Creating confusion matrix
knn = KNeighborsClassifier(n_neighbors=6, p = 2)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ["B", "M"])
cm_display.plot()
plt.show()
```

```python
n_pcs = pca.components_.shape[0]
most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]

initial_feature_names = X_train.columns
# get the names
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]

# LIST COMPREHENSION HERE AGAIN
dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}

# build the dataframe
df = pd.DataFrame(dic.items())
df
```
